```
def U1Dwire(Niter, N):
  U = np.zeros((N,N)) 
  U[0,:] = 100
  for n in np.arange(Niter):
    for i in np.arange(1,N-1):
      for j in np.arange(1,N-1):
        U[i,j] = 0.25*(U[i-1,j]+U[i+1,j]+U[i,j-1]+U[i,j+1])
  return U
```

```
N = 20
Niter = 50
U = U1Dwire(Niter, N)
```

```
plt.imshow(U)
plt.colorbar(label='U')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
